```python
# import jax.numpy as jnp
import numpy as jnp
import jax
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = [14,8]
plt.style.use('ggplot')
plt.rcParams['lines.markersize'] = 8
plt.rcParams.update({'font.size': 25})

import time
from tqdm import tqdm

from scipy.sparse import spdiags
```

```python
class Burgers:
    
    def __init__(self, viscosity, omega, Nx, Nt):
        
        self.Nx = Nx
        self.hx = 1/float(Nx)
        self.Nt = Nt
        self.ht = 1/float(Nt)
        
        self.viscosity = viscosity
        self.omega = omega
        self.prob_get()
        
    
    def prob_get(self):
        e = jnp.ones(self.Nx)
        # mass matrix
        self.M = (self.hx / 6.) * spdiags([e, 4*e, e], [-1, 0, 1], self.Nx-1, self.Nx-1).toarray()
        
        #sitffness matrix
        self.A = (self.viscosity/self.hx) * spdiags([-e, 2*e, -e], [-1, 0, 1], self.Nx-1, self.Nx-1).toarray()
        
        
        e = jnp.ones(self.Nx + 1)
        B = -(self.hx/6.) * spdiags([e, 4*e, e], [-1, 0, 1], self.Nx + 1, self.Nx + 1).toarray()
        self.B = B[1:self.Nx, :]
        
        self.Q = (self.hx/6.) * spdiags([e, 4*e, e], [-1, 0, 1], self.Nx+1, self.Nx+1).toarray()
        self.Q[0, 0] = self.hx / 3.
        self.Q[-1, -1] = self.hx / 3. 
        
        #vectors r on the right hand side of the Burgers equation
        # r[:, i] vector r at time step i + 1
        self.r = jnp.zeros((self.Nx - 1, self.Nt + 1))
        
        # vectors g in the objective function 
        # g[:, i] vector g at time step i + 1
        self.g = jnp.zeros((self.Nx - 1, self.Nt + 1))
        self.g[0:self.Nx//2, :] = - self.hx * jnp.ones((self.Nx//2, self.Nt + 1))
        
    def Ny(self, y):
        val = jnp.zeros(len(y))
        
        val[0] = y[0] * y[1] + y[1]**2
        for i in range(1, len(y) - 1):
            val[i] = -y[i-1]**2 - y[i] * y[i - 1] + y[i] * y[i + 1] + y[i + 1]**2
        val[-1] = -y[-2]**2 - y[-1]*y[-2]
        return val/6.
    
    def Nyp(self, y):
        
        # subdiagonal
        e1 = jnp.zeros(len(y))
        e1[:-1] = -2*y[:-1] - y[1:]
        
        # diagonal 
        e2 = jnp.zeros(len(y))
        e2[0] = y[1]
        e2[1:-1] = y[2:] - y[:-2]
        e2[-1] = -y[-2]
        
        # superdiagonal
        e3 = jnp.zeros(len(y))
        e3[1:] = y[:len(y) - 1] + 2*y[1:]
        
        mat = 1/6. * (spdiags([e1, e2, e3], [-1, 0, 1], len(y), len(y)))
        
        return mat
        
        
    
    def compute_state(self, u):
        
        iflag = False
        idebg = False
        
        nu = self.Nx + 1 # len(u) = (self.Nx + 1) * (self.Nt + 1)
        
        rhs = jnp.zeros(self.Nx - 1)
        rhs[:self.Nx//2] = jnp.ones(self.Nx//2)
        y = jnp.zeros((self.Nx - 1, self.Nt + 1))
        y[:, 0] = rhs
        
        for i in range(self.Nt - 1):
            # compute y[:, i+1] using Newton's Method
            tol = 1e-2 * min(self.ht**2, self.hx**2)
            max_iter = 10
            
            # initial guess
            y[:, i + 1] = y[:, i]
                        
            # residual (res0 is the residual component that is independent of y[:, i+1])
            res0 = (- self.M + self.ht/2. * self.A).dot(y[:, i]) + self.ht/2. * self.Ny(y[:, i]) \
                    + self.ht/2. * self.B.dot(u[(i)*nu:(i + 1)*nu] + u[(i + 1)*nu:(i + 2)*nu]) \
                    - self.ht/2. * (self.r[:, i] + self.r[:, i + 1])

            res  = res0 + (self.M + self.ht/2. * self.A).dot(y[:, i + 1]) + self.ht/2. * self.Ny(y[:, i + 1])
            
        iter_num = 0
        
        if idebg:
            print("")
            
        while iter_num < max_iter and jnp.linalg.norm(res) >= tol:
            resnorm = jnp.linalg.norm(res)
            
            # jacobian
            Mat = self.M + self.ht / 2. * (self.A + self.Nyp(y[:, i + 1]))
            
            # Newton Step
            stepy = - jnp.linalg.inv(Mat).dot(res)
            
            # Compute new guess using Armijo rules
            stepsize = 1
            ytmp = (y[:, i + 1] + stepsize*stepy).T
            res = res0 + (self.M + self.ht / 2. * self.A).dot(ytmp) + self.ht / 2. * self.Ny(ytmp)
            
            while jnp.linalg.norm(res) > (1 - 1e-4 * stepsize)*resnorm:
                stepsize = stepsize / 2. 
                ytmp = (y[:, i + 1] + stepsize * stepy).T
                res = res0 + (self.M + self.ht / 2. * self.A).dot(ytmp) + self.ht / 2. * self.Ny(ytmp)
            y[:, i + 1] = ytmp
            
            iter_num += 1 
            
        return y
    
Nt = 80 
Nx = 80
viscosity = 0.01
omega = 0.05 
burgers = Burgers(viscosity, omega, Nx, Nt)

u = jnp.zeros((Nt + 1)*(Nx + 1))
burgers.compute_state(u)
```

```python
class AirTrafficFlow():
    """Grid Layout is x-axis is spatial and y-axis is temporal. Hence we have:
    M[i, j] = (i h, j h). """
    def __init__(self, rho_max, L, T, v_min, v_max, h, rho_0):
        assert (int(L/h) == L/h) and (int(T/h) == T / h)
        assert len(rho_0) == L/h
        self.rho_max = rho_max
        self.L = L
        self.T = T
        self.h = h
        self.Ns = int(L / h)
        self.Nt = int(T / h)
        self.v_min = v_min
        self.v_max = v_max
        self.rho_0 = jnp.zeros(self.Nt*self.Ns)
        self.rho_0 = jax.ops.index_update(self.rho_0, jnp.array(range(self.Ns)), rho_0)

        
    def f(self, X):
#         dtRho =
        pass
        
    def _construct_M(self, X):
        h = self.h
        Nt = self.Nt
        Ns = self.Ns
        
        M = jnp.zeros((Nt * Ns, Nt * Ns))
        
        M = jax.ops.index_update(M, (jnp.array([range(Ns, (Nt - 2) * Ns)]), jnp.array([Ns*(i//Ns + 1) + i % Ns for i in range(Ns, (Nt - 2) * Ns)]),), 1./(2 * h))
        M = jax.ops.index_update(M, (jnp.array([range(Ns, (Nt - 2) * Ns)]), jnp.array([Ns*(i//Ns - 1) + i % Ns for i in range(Ns, (Nt - 2) * Ns)]),), -1./(2 * h))
        
        M = jax.ops.index_update(M, (jnp.array([1 + i + 2*(i//(Ns - 2)) for i in range(0, Nt * (Ns - 2))]), jnp.array([1 + i + 2*(i//(Ns - 2)) + 1 for i in range(0, Nt * (Ns - 2))]),), jnp.array([X[0][1 + i + 2*(i//(Ns - 2))]/(2. * h) for i in range(0, (Ns - 2) * Nt)]))
        M = jax.ops.index_update(M, (jnp.array([1 + i + 2*(i//(Ns - 2)) for i in range(0, Nt * (Ns - 2))]), jnp.array([1 + i + 2*(i//(Ns - 2)) - 1 for i in range(0, Nt * (Ns - 2))]),), jnp.array([-X[0][1 + i + 2*(i//(Ns - 2))]/(2. * h) for i in range(0, (Ns - 2) * Nt)]))
        
        M = jax.ops.index_update(M, (jnp.array([1 + i + 2*(i//(Ns - 2)) for i in range(0, Nt * (Ns - 2))]), jnp.array([1 + i + 2*(i//(Ns - 2)) for i in range(0, Nt * (Ns - 2))]),), jnp.array([(X[0][1 + i + 2*(i//(Ns - 2)) + 1] - X[0][1 + i + 2*(i//(Ns - 2)) - 1])/(2. * h) for i in range(0, (Ns - 2) * Nt)]))
        
        M = jax.ops.index_update(M, (jnp.array([i for i in range(Ns)]), jnp.array([Ns + i for i in range(0, Ns)]),), 1./(h))
        M = jax.ops.index_update(M, (jnp.array([i for i in range(Ns)]), jnp.array([i for i in range(0, Ns)]),), -1./(h))
        M = jax.ops.index_update(M, (jnp.array([Ns*(Nt - 1) + i for i in range(Ns)]), jnp.array([Ns*(Nt - 1) + i for i in range(0, Ns)]),), 1./(h))
        M = jax.ops.index_update(M, (jnp.array([Ns*(Nt - 1) + i for i in range(Ns)]), jnp.array([Ns*(Nt - 2) + i for i in range(0, Ns)]),), -1./(h))
        
        M = jax.ops.index_add(M, (jnp.array([i * Ns for i in range(Nt)]), jnp.array([Ns*i + 1 for i in range(Nt)]),),jnp.array([X[0][Ns*i]/(h) for i in range(Nt)]))
        M = jax.ops.index_add(M, (jnp.array([i * Ns for i in range(Nt)]), jnp.array([Ns*i for i in range(Nt)]),), jnp.array([-X[0][Ns*i]/(h) for i in range(Nt)]))
        M = jax.ops.index_add(M, (jnp.array([(i + 1)*Ns - 1 for i in range(Nt)]), jnp.array([(i + 1)*Ns - 1 for i in range(Nt)]),), jnp.array([X[0][(i + 1)*Ns - 1]/(h) for i in range(Nt)]))
        M = jax.ops.index_add(M, (jnp.array([(i + 1)*Ns - 1 for i in range(Nt)]), jnp.array([(i + 1)*Ns - 2 for i in range(Nt)]),), jnp.array([-X[0][(i + 1)*Ns - 1]/(h) for i in range(Nt)]))
        
        M = jax.ops.index_add(M, (jnp.array([i*Ns for i in range(Nt)]), jnp.array([Ns*i for i in range(Nt)]),), jnp.array([(X[0][Ns*i + 1] - X[0][Ns*i])/(h) for i in range(Nt)]))
        M = jax.ops.index_add(M, (jnp.array([(i + 1)*Ns - 1 for i in range(Nt)]), jnp.array([(i + 1)*Ns - 1 for i in range(Nt)]),), jnp.array([(X[0][(i + 1)*Ns - 1] - X[0][(i + 1)*Ns - 2])/(h) for i in range(Nt)]))

        return M
    
    def _solve_rho(self, X):
        M = self._construct_M(X)
        truncated_M = M[self.Ns:, self.Ns:]
        truncated_M_inv = jnp.linalg.inv(truncated_M)

        res = M.dot(self.rho_0)[self.Ns:]
#         print(M.dot(self.rho_0))
#         res = jax.ops.index_update(res, jnp.array(range(0, self.Ns)), M.dot(self.rho_0)[:self.Ns])
#         print(res)
        return truncated_M_inv.dot(-res)
        
```

```python
h = 0.05
L = 1
T = 1
v_min = None
v_max = None
rho_max = None
rho_0 = jnp.zeros(int(L/h))
rho_0 = jax.ops.index_update(rho_0, int(L/h * 0.25), 1)
ATF = AirTrafficFlow(rho_max, L, T, v_min, v_max, h, rho_0)
v = jnp.ones((1, int(L/h) * int(T/h)))

rho = ATF._solve_rho(v)
print(rho.shape)
rho
```
